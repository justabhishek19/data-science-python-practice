# Write a function to remove duplicates from a list

```python
def duplicates_remover(numbers):
    unique_nums = []
    for number in numbers:
        if number not in unique_nums:
            unique_nums.append(number)
    numbers = unique_nums
    return numbers                                    # O(n)
```

```python
numbers = [1,2,3,4,5,2,4]
duplicates_remover(numbers)
```
