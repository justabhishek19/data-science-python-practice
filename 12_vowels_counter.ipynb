# Write a function to count the number vowels and number of consonants in a string.
# NOTE: It contains spaces and special characters

```python
def count_vowels(string):
    v = ['a','e','i','o','u']
    vowels = 0
    consonants = 0
    for ch in string:
        if ch in v:
            vowels += 1
        elif ord(ch)>=ord('a') and ord(ch)<= ord('z'):
            consonants += 1
    return vowels, consonants
```

```python
s = '__hel lo 123$world '
vowels,consonants = count_vowels(s)
print(f'No. of vowels = {vowels}')
print(f'No. of consonants = {consonants}')
```

# Alternative CODE

```python
def count_vowels(string):
    vowels = 0
    consonants = 0
    for ch in string:
        if ch.isalpha():
            if ch in 'aeiou':
                vowels += 1
            else:
                consonants += 1
    return vowels, consonants
```

```python
s = '__hel lo 123$world '
vowels,consonants = count_vowels(s)
print(f'No. of vowels = {vowels}')
print(f'No. of consonants = {consonants}')
```
