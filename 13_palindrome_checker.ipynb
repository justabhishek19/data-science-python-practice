# Write a function to check if a string is palindrome or not?

```python
def is_palindrome(string):
    s = string.lower().replace(',', "").replace(" ", "").strip()

    i = 0
    j = len(s) - 1

    while i < j:
        if s[i] != s[j]:
            return False
        i += 1
        j -= 1

    return True
```

```python
is_palindrome('A man, a plan, a canal, Panama')
```

```python
is_palindrome('malayalam')
```

# Alternative Code

```python
def is_palindrome(string):
    s = string.lower().replace(',', "").replace(" ", "").strip()

    if s == s[::-1]:
        return True
    return False
```

```python
is_palindrome('A man, a plan, a canal, Panama')
```

```python
is_palindrome('malayalam')
```
