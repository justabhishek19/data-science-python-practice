# Write a Python function that takes a list of numbers as input and returns a new list containing only the even numbers from the original list.

```python
def evenList(numList):
    res = []
    for i in numList:
        if i%2==0:
            res.append(i)
    return res
```

```python
l1 = [i for i in range(11)]
l1
```

```python
even_nums = evenList(l1)
print(even_nums)
```
