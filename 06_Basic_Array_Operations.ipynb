```python
import array
```

```python
# creating

arr1 = array.array('i',[10,20,30,40,50])
print(arr1)
```

# Accessing (Indexing)

```python
arr1[3]
```

```python
arr1[0]
```

```python
arr1[4]
```

# Traversal

```python
for num in arr1:
    print(num,end=" ")
```

# Insertion

```python
arr1.insert(3,35)
arr1
```

# Deletion

```python
arr1.remove(30)  # Only removes first occurence 
arr1
```

# Updation

```python
arr1[2] = 30
arr1
```

# Sorting

```python
arr2 = array.array('i',[70,60,80])
arr2
```

```python
array.array('i',sorted(arr2))
```

# Merging

```python
arr1
```

```python
arr2
```

```python
arr3 = arr1 + arr2
arr3
```

# Splitting

```python
indx = int(len(arr3)/2)
```

```python
arr3[:indx]
```

```python
arr3[indx:]
```

```python
import numpy as np
```

```python
np1 = np.array([10,20,30,40,50])
np2 = np.array([50,60,70])
```

```python
# access
np1[3]
```

```python
# traversal
for item in np1:
    print(item,end=" ")
```

```python
# insertion
np.insert(np1,3,35)
```

```python
np.delete(np1,3)
```

```python
# updation
np1[2] = 69
np1
```

```python
# sorting
np3 = np.array([5,3,7,6,2,9])
np.sort(np3)
```

```python
np3
# that means it return copy not view
```

```python
np.concatenate((np1,np2))
```

```python
# splitting
split1, split2 = np.array_split(np3,2)
```

```python
split1
```
