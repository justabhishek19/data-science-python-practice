# Write a function to check if two strings are anagrams or not.
# NOTE: Two string are anagram if both has same characters with the same frequency.

```python
def is_anagram(s1,s2):
    if sorted(s1) == sorted(s2):
        return True
    return False
```

```python
s1 = 'listen'
s2 = 'silent'
is_anagram(s1,s2)
```

# Alternative Code

```python
def is_anagram(s1,s2):
    alpha_indx = [0 for i in range(26)]

    for ch in s1:
        alpha_indx[ord(ch)-97] += 1
        
    for ch in s2:
        alpha_indx[ord(ch)-97] -= 1
        
    if  all([x == 0 for x in alpha_indx]):
        return True
        
    return False
```

```python
s1 = 'listen'
s2 = 'silent'
is_anagram(s1,s2)
```
