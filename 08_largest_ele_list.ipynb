# Write a function to find out the largest element present in a list

```python
def max_ele(arr):
    maxx = arr[0]
    for num in arr:
        if num > maxx:
            maxx = num
    return maxx                               # O(n)
```

```python
numbers = [-2,9,100,-199,45,98,47]
max_ele(numbers)
```
